```python
dataBase = mysql.connector.connect( 
  host ="localhost", 
  user ="root", 
  passwd ="password",
  database = "College"
) 
```

```python
# preparing a cur object 
cur = dataBase.cursor()
```

```python
query = """
    CREATE TABLE IF NOT EXISTS books(
        Book_id INTEGER(10) PRIMARY KEY, 
        Title varchar(20) NOT NULL,
        Author varchar(20),
        Price INTEGER,
        Available varchar(10)
    )
"""
cur.execute(query)
```

```python
pd.read_sql("desc books",dataBase)
```

```python
query = """
    CREATE TABLE IF NOT EXISTS Members(
        Member_id INTEGER(10) PRIMARY KEY, 
        Name varchar(20) NOT NULL,
        City varchar(20),
        Joining_Date DATE
    )
"""
cur.execute(query)
```

```python
pd.read_sql("desc members",dataBase)
```

```python
query = """
    CREATE TABLE IF NOT EXISTS Borrow(
        Book_id INTEGER(10) REFERENCES books (book_id),
        Member_id INTEGER(10) REFERENCES members (member_id),
        Issue_Date DATE,
        Return_Date DATE
    )
"""
cur.execute(query)
```

```python
pd.read_sql("desc borrow",dataBase)
```

```python
query = """
    CREATE TABLE IF NOT EXISTS Publisher(
        Publisher_id INTEGER(10) PRIMARY KEY,
        Name varchar(10),
        Book_id INTEGER(10) REFERENCES books (book_id)
    )
"""
cur.execute(query)
```

```python
pd.read_sql("desc publisher",dataBase)
```
